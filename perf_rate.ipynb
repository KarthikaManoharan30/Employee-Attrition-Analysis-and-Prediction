```python
# Binary encoding for Attrition, Gender, Over18, and OverTime
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x =='Yes'else 0)
df['Gender'] = df['Gender'].apply(lambda x: 1 if x =='Male'else 0)
df['Over18'] = df['Over18'].apply(lambda x: 1 if x =='Y'else 0)
df['OverTime'] = df['OverTime'].apply(lambda x: 1 if x =='Yes'else 0)
```

```python
df
```

```python
# OneHot Encoding for BusinessTravel', 'Department', 'EducationField' 'JobRole', 'MaritalStatus'

cat_cols = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']

df = pd.get_dummies(df, columns=cat_cols, prefix=cat_cols, drop_first=False)

# Convert True/False to 1/0
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
# Boxplots: Numeric features vs PerformanceRating
features = ['Age', 'JobLevel', 'MonthlyIncome', 'PercentSalaryHike', 'YearsAtCompany', 'JobInvolvement', 'YearsInCurrentRole' ]

for feature in features:
    plt.figure(figsize=(6, 4))
    custom_palette = {3: 'brown', 4: 'pink'}  # colors for the two ratings
    sns.boxplot(x='PerformanceRating', y=feature, data=df, hue='PerformanceRating', palette=custom_palette, legend=False)
    plt.title(f'{feature} vs. Performance Rating')
    plt.xticks([0, 1], ['Rating 3', 'Rating 4'])  # custom labels
    plt.show()
```

```python
# These columns are constant or unique IDs → not useful for prediction

drop_cols = ["EmployeeCount", "EmployeeNumber", "Over18", "StandardHours"]

df = df.drop(columns=drop_cols)
print("Remaining columns:", df.columns.tolist())
```

```python
# Correlation heatmap after preprocessing


# Now recalculate correlation
corr = df.corr(numeric_only=True)

# Heatmap
plt.figure(figsize=(16,10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Top correlations with PerformanceRating
corr_target = corr['PerformanceRating'].drop('PerformanceRating').sort_values(ascending=False)
print("Top Positive Correlations:\n", corr_target.head())
print("\nTop Negative Correlations:\n", corr_target.tail())
```

```python
# Machine Learning Model

# Split the data into features and target variable

x = df.drop(columns=['PerformanceRating'])
y = df['PerformanceRating']
```

```python
# Train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Choose a model (logistic regression)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)

```

```python
lr_score = lr.score(x_test, y_test)
print(f"Model Score: {lr_score:.4f}")
```

```python
# Confusion Matrix and Classification Report
from sklearn.metrics import confusion_matrix, classification_report
y_pred = lr.predict(x_test)
cm = confusion_matrix(y_test, y_pred)


import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Rating 3', 'Rating 4'], yticklabels=['Rating 3', 'Rating 4'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print("Classification Report:\n", classification_report(y_test, y_pred, target_names=['Rating 3', 'Rating 4']))

```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

lr_pipeline = Pipeline(steps=[
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression(
        solver='lbfgs',
        max_iter=1000,
        class_weight='balanced'
    ))
])
```

```python
lr_pipeline.fit(x_train, y_train)
```

```python
y_pred = lr_pipeline.predict(x_test)
```

```python
# Accuracy score, confusion matrix, and classification report
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=lr_pipeline.classes_)
disp.plot(cmap='Reds')
plt.title("Confusion Matrix - Performance Rating Prediction")
plt.show()


print("\nClassification Report:\n", classification_report(y_test, y_pred))

```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Get predicted probabilities for the positive class (Rating 4)
y_prob = lr_pipeline.predict_proba(x_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob, pos_label=4)
roc_auc = roc_auc_score(y_test, y_prob)

# Plot
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Performance Rating Prediction')
plt.legend(loc='lower right')
plt.show()

       
```
